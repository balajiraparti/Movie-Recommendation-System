```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words=set(stopwords.words('english'))
lemmatizer=WordNetLemmatizer()
```

```python
import string
def remove_punctuation(text:str):
    text=str(text).lower()
    text=re.sub(r'[^a-zA-Z\s]','',text)
    words=text.split()
    words=[word for word in words if word not in stop_words]
    words=[lemmatizer.lemmatize(word) for word in words]
    return ' '.join(words)

df_new['tags']=df_new['tags'].apply(remove_punctuation)
```

```python
df_new=df_new.reset_index(drop=True)
```

```python
indices=pd.Series(df_new.index,index=df_new['title']).drop_duplicates()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer(ngram_range=(1,2),max_features=50000,stop_words='english')
```

```python
tfidf_matrix=tfidf.fit_transform(df_new['tags'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
type(indices)
```

```python
def recommend(title,n=10):
    if title not in indices:
        return ['Movie not found']
    idx=indices[title]
    sim_score=cosine_similarity(tfidf_matrix[idx],tfidf_matrix).flatten()
    similar_idx=sim_score.argsort()[::-1][1:n+1]
    print(similar_idx)
    return df_new.iloc[similar_idx]

    
    
```

```python
recommend("Interstellar")
```
